```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1])
```

```python
df = pd.DataFrame(x)
x0 =df[0].values
x1 = df[1].values
```

```python

plt.scatter(x0[:3], x1[:3] , marker='o', label='y=0')
plt.scatter(x0[3:], x1[3:] , marker='x', label='y=1',c='r')
plt.axis([0,3.5,0,3])
plt.xlabel("$x_0$")
plt.ylabel("$x_1$")
plt.legend()
plt.show()
```

```python
x = np.arange(0,6)

y = 3 - x
plt.scatter(x0[:3], x1[:3] , marker='o', label='y=0')
plt.scatter(x0[3:], x1[3:] , marker='x', label='y=1',c='r')
plt.axis([0,4,0,4])
plt.plot(x,y,linewidth=5 ,c='g')
plt.fill_between(x,y,alpha =0.2 ) # filling with color and alpha is for transparency
plt.xlabel("$x_0$")
plt.ylabel("$x_1$")
plt.legend()
plt.show()
```
