```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
z = np.arange(-10,11)
```

```python
def sigmoid(z):
    exp = np.exp(-z) # it will return scaler for scaler input and vector for vector input
    g = 1/(1+exp)
        
    return g  # Returns:
                    #g (ndarray): sigmoid(z), with the same shape as z
```

```python
np.set_printoptions(precision =3)
y = sigmoid(z)

print(np.c_[z, y])
```

```python
plt.plot(z , sigmoid(z) , 'r--')
```

```python
def reg(x , w ,b):
    return np.dot(x,w) +b
```

```python
x = np.array([0., 1, 2, 3, 4, 5])
y = np.array([0,0,0,1,1,1])
z = reg (x, 4.54, -11.15)

ys = sigmoid(z)

# plt.plot(x,ys ,'r')
plt.plot(x,ys)
plt.scatter(x,y, marker ='*' ,c='r')
#plt.plot(x,z)
```
