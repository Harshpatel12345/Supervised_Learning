```python
import numpy as np 
import matplotlib.pyplot as %plt
import pandas as pd
```

```python
x_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])  #(m,n)
y_train = np.array([0, 0, 0, 1, 1, 1])                                           #(m,)
```

```python
df = pd.DataFrame(x_train)
x0 = df[0].values
x1 = df[1].values
```

```python
plt.scatter(x0[:3],x1[:3], label='y=0', c ='g')
plt.scatter(x0[3:],x1[3:], label='y=1', c='r' ,marker ="X")
plt.axis([0,3.5,0,3])
plt.legend()
                            #alpha : float, default: None
                         #The alpha blending value, between 0 (transparent) and 1 (opaque).
plt.xlabel("$X_0$------>",fontsize=15)
plt.ylabel("$X_1$------>",fontsize=15)
plt.title("Categorical points for tumur prediction")
plt.show()
```

```python
def sigmoid(z):
    exp = np.exp(-z)
    g =1/(1+exp)   # g = 1/(1+e^(-z))
    return g
```

```python
def cost(x,y,w,b):
    m = x.shape[0]
    cost = 0.0
    for i in range(m):
        z = np.dot(x[i],w) + b
        f = sigmoid(z) # f(w.x +b)= g(z) 
        loss = -y[i]*np.log(f) -(1-y[i])*np.log(1-f)
        cost+=loss
        
    cost/=m
    
    return cost
```

```python
w_in =[1,1]
b_in =-3 

print(cost(x_train,y_train,w_in,b_in))
```

```python
w_in =[1,1]
b_in =-3

print("Cost for b = -3:",cost(x_train,y_train,w_in,b_in))
print("Cost for b = -4:",cost(x_train,y_train,w_in,-4))
```
