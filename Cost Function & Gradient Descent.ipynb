```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])   #features
y_train = np.array([300.0, 500.0])
```

```python
plt.scatter(x_train, y_train ,marker ='*' )
plt.title("House Price")
plt.xlabel("Area")

plt.ylabel("Price")
plt.show()
```

```python
def cost_fun(x,y,w,b):
    m = len(x)
    cost =0
    for i in range(m):
        f_es = w*x[i] +b
        cost = cost + (f_es - y[i])**2
        
    total_cost = (1 /(2*m))*cost
    
    return total_cost
        
```

```python
w = []
jw = []
for i in range(100,350,5):
    w.append(i) # Storing all the value of w for plotting of jw
    jw.append(cost_fun(x_train,y_train ,i,50)) # calculating cost function for a single w and storing in list
```

```python
plt.plot(w,jw ,'r-')
plt.title("Cost Function")
plt.ylabel("J(w,b)-------->")
plt.xlabel("w------>")
plt.show()
```

```python
def model(x,w,b):
    m=len(x)
    f=[]
    for i in range(m):
        f.append(w*x[i] +b)
        
    return f
```

```python
f_es = model(x_train,225,50)

plt.plot(x_train, f_es , 'g*--' ,label ="Our prediction")
plt.plot(x_train,y_train , c ='r' , label = "Actual Plot")
plt.legend()
plt.title("Our Prediction Graph")
plt.xlabel("size------->")
plt.ylabel("Price------->")
plt.show()
```

# Gradient Descent for Linear Regression

```python
def compute_gradient(x,y,w,b):
    m = len(x)
    sum_w =0
    sum_b =0
    for i in range(m):
        sum_w+=(w*x[i] +b -y[i])*x[i]
        sum_b +=(w*x[i] +b - y[i])
    dw = sum_w/m
    db = sum_b/m
    
    return dw,db
        
```

```python
alpha = 1.0e-3
j =[]
parm =[]
def gradient_descent(w,b):
    for i in range(20000):
        dw, db = compute_gradient(x_train, y_train,w,b)
        
        w = w- alpha*dw
        b = b- alpha*db
        j.append(cost_fun(x_train,y_train,w,b))
        parm.append([w,b])
    return w,b 
```
